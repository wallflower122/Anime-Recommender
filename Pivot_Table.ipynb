```python
import re
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

anime_feature['name'] = anime_feature['name'].apply(text_cleaning)
```

```python
anime_pivot=anime_feature.pivot_table(index='name',columns='user_id',values='user_rating')
anime_pivot.head()
```
